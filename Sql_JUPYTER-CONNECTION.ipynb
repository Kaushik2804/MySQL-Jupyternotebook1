```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
user = 'root'
password = 'root'
host = 'localhost'
db_name = 'first_db'
port = 3306
```

```python
sqlEngine = create_engine(f'mysql+pymysql://{user}:{password}@{host}/{db_name}', pool_recycle=port)
dbConnection = sqlEngine.connect()
```

```python
sql_query = "select * from first_db.formatted_microsoft"
```

```python
df = pd.read_sql(sql_query, dbConnection)
```
